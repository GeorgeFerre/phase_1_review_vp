# Task
> create a dictionary of where each key is a market, and each value is a list of songs available in that market

```python
# Fill in None with the appropriate value
market_dict = {}

for market in markets:
    market_dict[market] = []
    
for market in markets:
    for song in songs:
        if market in song['available_markets']:
            market_dict[market].append(song['name'])
            
market_dict
# your code here
```

# Task
Write a function which takes a market and a song_name as arguments, and returns the song name if the song is in the market.

```python
# Your code here
def song_not_in_market(market, song_name):
    if song_name not in market_dict[market]:
        return song_name
```
